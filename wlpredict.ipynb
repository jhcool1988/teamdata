```
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from scipy.stats import poisson,skellam
from scipy import stats
import sklearn.linear_model
```

```
epl_result = pd.read_csv('https://raw.githubusercontent.com/ljg-95/bigdataset1/master/results.csv')
epl_result

```

```
epl_result = epl_result[['home_team','away_team','home_score','away_score']]
epl_result.head()
print(epl_result.mean())
# 평균적으로 홈팀이 원정팀보다 골을 더 많이 넣는다
```

```
skellam.pmf(0.0,  epl_result.mean()[0],  epl_result.mean()[1])
# 홈팀과 원정팀이 비길 확률
```

```
skellam.pmf(1,  epl_result.mean()[0],  epl_result.mean()[1])
# 한골로 홈팀이 승리할 확률
```

```
# 포아송회귀모형에 필요한 툴
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

```
goal_model_data = pd.concat([epl_result[['home_team','away_team','home_score']].assign(home=1).rename(
            columns={'home_team':'team', 'away_team':'opponent','home_score':'goals'}),
           epl_result[['away_team','home_team','away_score']].assign(home=0).rename(
            columns={'away_team':'team', 'home_team':'opponent','away_score':'goals'})])

poisson_model = smf.glm(formula="goals ~ home + team + opponent", data=goal_model_data, 
                        family=sm.families.Poisson()).fit()
poisson_model.summary()
```

```
poisson_model.predict(pd.DataFrame(data={'team': 'Singapore', 'opponent': 'Zambia',
                                       'home':1},index=[1]))
```

```
poisson_model.predict(pd.DataFrame(data={'team': 'Zambia', 'opponent': 'Singapore',
                                       'home':0},index=[1]))
```

```
def simulate_match(foot_model, homeTeam, awayTeam, max_goals=10):
    home_goals_avg = foot_model.predict(pd.DataFrame(data={'team': homeTeam, 
                                                            'opponent': awayTeam,'home':1},
                                                      index=[1])).values[0]
    away_goals_avg = foot_model.predict(pd.DataFrame(data={'team': awayTeam, 
                                                            'opponent': homeTeam,'home':0},
                                                      index=[1])).values[0]
    team_pred = [[poisson.pmf(i, team_avg) for i in range(0, max_goals+1)] for team_avg in [home_goals_avg, away_goals_avg]]
    return(np.outer(np.array(team_pred[0]), np.array(team_pred[1])))
simulate_match(poisson_model, 'Singapore', 'Zambia', max_goals=3)
```

```
# Singapore win
Singapore_win = simulate_match(poisson_model, "Singapore", "Zambia", max_goals=10)
np.sum(np.tril(Singapore_win, -1))
```

```
# draw
np.sum(np.diag(Singapore_win))
```

```
# Zambia win
np.sum(np.triu(Singapore_win, 1))
```
