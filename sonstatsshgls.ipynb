```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
```

```
df = pd.read_csv('https://raw.githubusercontent.com/jhcool1988/teamdata/master/son.csv')
df.head()

print(df)

plt.figure(figsize=(18, 9))

plt.plot(range(shgls.shape[0]), shgls, color='r')

plt.show()
```

```
shgls = df.loc[:, 'shgls'].values

shgls.shape
```

```
train_data = shgls[:200]
test_data = shgls[200:321]

train_data = train_data.reshape(-1, 1)
test_data = test_data.reshape(-1, 1)

print(train_data)
print(test_data)

print('%d training and %d total testing instances' % (len(train_data), len(test_data)))
```

```
plt.subplot(1, 2, 1)
plt.plot(range(train_data.shape[0]), train_data, color='r', label='Training split')

plt.subplot(1, 2, 2)
plt.plot(range(test_data.shape[0]), test_data, color='b', label='Test Split')

plt.tight_layout()
plt.show()
```

```
# 윈도우 표준화
normalization_window = 50

scaler = MinMaxScaler(feature_range=(0, 1))

for i in range(0, 200, normalization_window):
    scaler.fit(train_data[i:i + normalization_window, :])
    train_data[i:i + normalization_window, :] = scaler.transform(train_data[i:i + normalization_window, :])

test_data = scaler.fit_transform(test_data)

print(test_data)
```

```
# 지수 평활법 구현
Smoothing = 0.0
gamma = 0.1

for i in range(200):
    Smoothing = gamma * train_data[i] + (1 - gamma) * Smoothing
    train_data[i] = Smoothing
```

```
# 단순이동예측
from sklearn.metrics import mean_squared_error

window_size = 15     # ?
N = train_data.size

std_avg_predictions=[]
mse_errors = []

for i in range(window_size, N):
  std_avg_predictions.append(np.mean(train_data[i - window_size:i]))
  mse_errors.append((std_avg_predictions[-1] - train_data[i]) ** 2)

print('MSE error for standard averaging: %.5f' % (0.5 * np.mean(mse_errors)))
```

```
# 표준 이동 평균 예측 결과 플롯
plt.figure(figsize=(19,6))
plt.plot(range(train_data.shape[0]), train_data,
               color='darkblue', label='Actual')
plt.plot(range(window_size, N), std_avg_predictions, color='orange', label='Predicted')
plt.xticks(range(0, df.shape[0] - len(test_data), 50),
           df['idx'].loc[::50], rotation=45)
plt.xlabel('idx')
plt.ylabel('shgls')
plt.legend(fontsize=18)
plt.show()
```

```
# 지수적 이동 평균 계산
ema_avg_predictions = []
mse_errors = []

EMA = 0.0
ema_avg_predictions.append(EMA)

gamma = 0.5
window_size = 100
N = len(train_data)

for i in range(1, N):
  EMA = EMA * gamma + (1.0 - gamma) * train_data[i - 1]
  ema_avg_predictions.append(EMA)
  mse_errors.append((ema_avg_predictions[-1] - train_data[i]) **2 )

print('MSE eooro for EMA averaging: %.5f' % (0.5 * np.mean(mse_errors)))
```

```
# 지수 이동 평균(EMA) 예측값 플롯
plt.figure(figsize=(19, 6))
plt.plot(range(train_data.shape[0]), train_data, color='darkblue', label='True')
plt.plot(range(0, N), ema_avg_predictions, color='orange', label='Prediction')
plt.xticks(range(0, df.shape[0] - len(test_data), 50),
           df['idx'].loc[::50], rotation=45)

plt.xlabel['idx']
plt.ylabel['shgls']
plt.legend(fontsize=18)
plt.show()
```
